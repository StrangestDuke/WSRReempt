```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../db/ComputerSales.csv')
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
"""

Distribution plots

"""
```

```python
sns.distplot(crash_df['not_distracted'], bins=20)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df,kind='reg')
plt.show()
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.pairplot(crash_df)

plt.figure(dpi=200)
plt.show()

plt.savefig('foo.png')
```

```python
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue='sex', palette='Blues')
plt.show()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8,4)) 

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df,estimator = np.std)
```

```python
sns.countplot(x='sex', data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips_df, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips_df,jitter=True, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips_df,hue='sex')
```

```python
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
```
